```python
# 라이브러리를 불러옵니다.
import numpy as np
import pandas as pd
```

```python
# 데이터를 불러옵니다.
data = pd.read_csv('raw_data/건강보험심사평가원_ATC코드 매핑 목록_20240630.csv', encoding='cp949')
print(data.info())
data.head()
```

```python
# NaN 값이 있는 열을 제거합니다.
data = data.dropna()
```

```python
# 항우울제 카테고리 및 처방약물 목록을 정의합니다.
drug_categories = {
    "Selective serotonin reuptake inhibitors (SSRIs)": ["citalopram", "escitalopram", "fluoxetine", "fluvoxamine", "paroxetine", "sertraline", "vilazodone"],
    "Norepinephrine-dopamine reuptake inhibitors": ["bupropion"],
    "Serotonin modulators (5-HT2 blockers)": ["mirtazapine", "trazodone"],
    "Serotonin-norepinephrine reuptake inhibitors": ["desvenlafaxine", "duloxetine", "levomilnacipran", "venlafaxine"],
    "Heterocyclic (including tricyclic) antidepressants": ["amitriptyline", "amoxapine", "clomipramine", "desipramine", "doxepin", "imipramine", "maprotiline", "nortriptyline", "protriptyline", "trimipramine"],
    "Monoamine oxidase inhibitors (MAOIs)": ["tranylcypromine", "phenelzine", "selegiline", "isocarboxazid"],
    "Psychostimulants": ["dextroamphetamine", "methylphenidate"],
    "Melatonergic antidepressant": ["agomelatine"],
    "Ketamine and esketamine": ["ketamine", "esketamine"]
}
```

```python
psychotropic_drugs = {
    "Methylphenidate": ["methylphenidate"],
    "Benzodiazipine": ["alprazolam", "chlordiazepoxide", "clorazepate", "clonazepam", "diazepam", "flurazepam", "lorazepam", "oxazepam", "temazepam", "triazolam"],
    "Zolpidem": ["zolpidem"],
}
```

```python
# 딕셔너리에 포함된 약물만 추출해서 새로운 데이터프레임을 만듭니다.
df_antidepressant = pd.DataFrame()
for category, drugs in drug_categories.items():
    for drug in drugs:
        df_antidepressant = pd.concat([df_antidepressant, data[data['ATC코드 명칭'].str.contains(drug, case=False)]])

df_antidepressant
```

```python
# 딕셔너리에 포함된 약물만 추출해서 새로운 데이터프레임을 만듭니다.
df_psychotropic = pd.DataFrame()
for category, drugs in psychotropic_drugs.items():
    for drug in drugs:
        df_psychotropic = pd.concat([df_psychotropic, data[data['ATC코드 명칭'].str.contains(drug, case=False)]])

df_psychotropic
```

```python
# 주성분코드, ATC코드, ATC코드 명칭을 매핑한 데이터프레임을 만듭니다.
df_mapping = data[['주성분코드', 'ATC코드', 'ATC코드 명칭']].drop_duplicates().reset_index(drop=True)
df_mapping.rename(columns={'ATC코드 명칭': '주성분명'}, inplace=True)
df_mapping.to_csv('data/name_code_mapping.csv', index=False)
```

```python
print(df_mapping['주성분명'].unique())
```

```python
# 주성분코드와 ATC코드 리스트를 정의합니다.
ingredient_code = list(df_antidepressant['주성분코드'].drop_duplicates())
ATC_code = list(df_antidepressant['ATC코드'].drop_duplicates())
```

```python
# 주성분코드와 ATC코드 리스트를 txt 파일로 저장합니다.
with open('data/ingredient_code.txt', 'w') as f:
    for code in ingredient_code:
        f.write(code + '\n')

with open('data/ATC_code.txt', 'w') as f:
    for code in ATC_code:
        f.write(code + '\n')
```
